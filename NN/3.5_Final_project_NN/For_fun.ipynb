```python
import tensorflow as tf
from keras.models import load_model
from matplotlib import pyplot as plt
import numpy as np
```

```python
test_ds = tf.keras.utils.image_dataset_from_directory(
    './little_test',
    image_size = (128, 128)
)
```

```python
model = load_model('best_model_conv.keras')
```

```python
class_labels = test_ds.class_names
iterator = test_ds.as_numpy_iterator()
image, label = next(iterator)
predictions = model.predict(image)

dct = {
    0: 'chihuahua',
    1: 'muffin'   
}

num_images = 11
num_images_per_row = 4  # Set the number of images per row
num_rows = 3

plt.figure(figsize=(15, 10))

for i in range(num_images):
    
    number_of_row = i // num_images_per_row

    predicted_class = class_labels[0] if predictions[i][0] < 0.5 else class_labels[1]
    true_class = dct[label[i]]

    plt.subplot(num_rows, num_images_per_row, i + 1)
    plt.imshow(image[i].astype("uint8"))
    plt.title(f'True: {true_class}\nPredicted: {predicted_class}, {predictions[i]}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
